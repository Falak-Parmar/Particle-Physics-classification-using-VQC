# Experiment 02 — Feature Scaling (BCE Standard)

**Goal:** Investigate VQC scaling from 2 to 8 features using Binary Cross-Entropy loss and the corrected data pipeline.

```python
import sys
sys.path.append('..')
import numpy as np
import pennylane as qml
from pennylane import numpy as pnp
from sklearn.metrics import roc_auc_score
from utils.data_utils import load_higgs, HIGGS_FEATURE_RANKING

def train_scaling(n_feat):
    X_train, X_val, X_test, y_train, y_val, y_test = load_higgs(
        path='../data/HIGGS.csv.gz', n_samples=1000, n_features=n_feat, scale_range=(0, np.pi)
    )
    
    dev = qml.device('lightning.qubit', wires=n_feat)
    
    @qml.qnode(dev, interface='autograd')
    def circuit(weights, x):
        for i in range(n_feat): qml.RY(x[i], wires=i)
        for l in range(2):
            for q in range(n_feat): qml.Rot(*weights[l, q], wires=q)
            for q in range(n_feat): qml.CNOT(wires=[q, (q + 1) % n_feat])
        return qml.expval(qml.PauliZ(0))
    
    def vqc_prob(w, x, b):
        return (circuit(w, x) + b + 1.0) / 2.0
    
    def bce_loss(weights, bias, X, y):
        probs = pnp.array([vqc_prob(weights, x, bias) for x in X])
        probs = pnp.clip(probs, 1e-15, 1.0 - 1e-15)
        return -pnp.mean(y * pnp.log(probs) + (1 - y) * pnp.log(1 - probs))
    
    pnp.random.seed(42)
    w = pnp.array(pnp.random.uniform(0, 2*np.pi, (2, n_feat, 3)), requires_grad=True)
    b = pnp.array(0.0, requires_grad=True)
    opt = qml.AdamOptimizer(stepsize=0.05)
    
    for epoch in range(20):
        idx = np.random.permutation(len(X_train))
        for start in range(0, len(X_train), 32):
            Xb, yb = X_train[idx][start:start+32], y_train[idx][start:start+32].astype(float)
            w, b = opt.step(lambda w_, b_: bce_loss(w_, b_, Xb, yb), w, b)
            
    test_probs = np.array([float(vqc_prob(w, x, b)) for x in X_test])
    return roc_auc_score(y_test, test_probs)

for nf in [2, 4, 8]:
    auc = train_scaling(nf)
    print(f"Features: {nf} | Test AUC: {auc:.4f}")
```
